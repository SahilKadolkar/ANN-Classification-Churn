```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
from tensorflow.keras.models import load_model
```

```python
model=load_model('model.h5')

with open('onehot_encoder_geo.pkl','rb') as file:
    label_encoder_geo=pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file) 
```

```python
input_data={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
geo_encoded=label_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encoder_df=pd.DataFrame(geo_encoded,columns=label_encoder_geo.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
print(label_encoder_gender.classes_)
```

```python
input_data=pd.concat([input_data.reset_index(drop=True),geo_encoder_df],axis=1)
```

```python
input_df['Gender'] = input_df['Gender'].map({'Male': 1, 'Female': 0})
input_df
```

```python
input_df=pd.concat([input_df.drop("Geography",axis=1),geo_encoder_df],axis=1)
input_df
```

```python
input_scaled=scaler.transform(input_df)
```

```python
prediction=model.predict(input_scaled)
```
